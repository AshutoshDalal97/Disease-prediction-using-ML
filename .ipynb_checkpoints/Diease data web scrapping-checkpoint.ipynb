```python
def extract_symptoms_clean(url):
    import requests
    from bs4 import BeautifulSoup
    
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")

    symptoms = []

    # Look for headings containing "symptoms"
    for tag in soup.find_all(["h2", "h3"]):
        if "symptom" in tag.text.lower():
            
            # Get next elements (lists or paragraphs)
            next_node = tag.find_next_sibling()
            
            while next_node and next_node.name in ["ul", "p"]:
                
                if next_node.name == "ul":
                    for li in next_node.find_all("li"):
                        symptoms.append(li.text.strip())
                
                elif next_node.name == "p":
                    symptoms.append(next_node.text.strip())
                
                next_node = next_node.find_next_sibling()
    
    return symptoms
```

```python
import re

def clean_symptoms(symptoms):
    cleaned = []
    
    for s in symptoms:
        s = s.lower()
        
        # remove brackets
        s = re.sub(r"\(.*?\)", "", s)
        
        # split by commas
        parts = re.split(r",|and", s)
        
        for p in parts:
            p = p.strip()
            
            # keep meaningful phrases only
            if len(p) > 3 and not p.startswith(("your", "the", "a ")):
                cleaned.append(p)
    
    return list(set(cleaned))
```

```python
def normalize_phrase(p):
    p = p.replace(" ", "_")
    
    # remove junk phrases
    junk = ["you", "your", "doctor", "will", "may"]
    if any(j in p for j in junk):
        return None
    
    return p
```

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import quote

def get_symptoms(disease):
    try:
        url = "https://en.wikipedia.org/wiki/" + quote(disease)

        headers = {"User-Agent": "Mozilla/5.0"}
        res = requests.get(url, headers=headers)
        soup = BeautifulSoup(res.text, "lxml")

        infobox = soup.find("table", {"class": "infobox"})

        if infobox:
            for row in infobox.find_all("tr"):
                header = row.find("th")

                if header and "symptom" in header.text.lower():
                    return row.find("td").text.strip()

        return None

    except:
        return None
```

```python
import pandas as pd
import time

data = []

for disease in diseases:
    print("Scraping:", disease)

    symptoms = get_symptoms(disease)

    if symptoms:
        data.append([disease, symptoms])

    time.sleep(1)

df = pd.DataFrame(data, columns=["Disease", "Symptoms"])
print(df.head())
```

```python
import re

def clean_text(text):
    text = re.sub(r'\[.*?\]', '', text)
    text = text.lower()
    text = text.replace("\n", ",")
    return text

def split_symptoms(text):
    parts = re.split(r',|;|and', text)
    return [p.strip() for p in parts if len(p.strip()) > 2]

def normalize(sym):
    return sym.replace(" ", "_")

df["Symptoms"] = df["Symptoms"].apply(clean_text)
df["Symptoms"] = df["Symptoms"].apply(split_symptoms)
df["Symptoms"] = df["Symptoms"].apply(
    lambda x: [normalize(s) for s in x]
)
```

```python
def filter_symptoms(sym_list):
    bad_words = ["may", "can", "include", "often", "usually"]

    return [s for s in sym_list if not any(b in s for b in bad_words)]

df["Symptoms"] = df["Symptoms"].apply(filter_symptoms)
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
import pandas as pd

mlb = MultiLabelBinarizer()

X = mlb.fit_transform(df["Symptoms"])

final_df = pd.DataFrame(X, columns=mlb.classes_)

final_df["prognosis"] = df["Disease"]
```

```python
final_df.to_csv("disease_dataset.csv", index=False)

print("✅ Dataset ready!")
```
