```python
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score, confusion_matrix,classification_report
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier,plot_tree
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Training.csv")
df=df.drop_duplicates()
df1=pd.read_csv("Testing.csv")
df=df.iloc[:,0:133]
df1=df1.iloc[:,0:133]
print(df.shape)
print(df.head())
```

```python
df.columns
```

```python
print(df['prognosis'].value_counts())
```

```python
X = df.drop('prognosis', axis=1)
y = df['prognosis']
X_train = df.drop('prognosis', axis=1)
y_train = df['prognosis']
X_test = df1.drop('prognosis', axis=1)
y_test = df1['prognosis']
#X_train, X_test, y_train, y_test = train_test_split(
 #   X, y, test_size=0.2, random_state=11,stratify=y)
```

```python
from sklearn.preprocessing import LabelEncoder 
le=LabelEncoder()
y_train_numeric=le.fit_transform(y_train)
y_test_numeric=le.transform(y_test)
print(y_train_numeric)
print(y_test_numeric)
print(le.classes_)
```

```python
def calculate_accuracy(y_test,y_pred):
    print("\n Confusion Matrix \n", confusion_matrix(y_test,y_pred))
    print("\n Accuracy Score \n", accuracy_score(y_test, y_pred) * 100)
    print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

# Model DT

```python
model_dt = DecisionTreeClassifier(
    criterion='gini', 
    max_depth=None,  # Allow tree to grow fully
    min_samples_split=5,
    min_samples_leaf=2,
    random_state=11
)
model_dt.fit(X_train, y_train)
y_pred_dt=model_dt.predict(X_test)
```

# Model RF

```python
from sklearn.ensemble import RandomForestClassifier
model_rf=RandomForestClassifier(n_estimators=100,criterion='gini',random_state=11,class_weight='balanced')
model_rf.fit(X_train,y_train)
y_pred_rf=model_rf.predict(X_test)
```

# Extremely Randomized Trees Implementation 

```python
from sklearn.ensemble import ExtraTreesClassifier

et_model = ExtraTreesClassifier(
    n_estimators=100,
    random_state=11
)

et_model.fit(X_train, y_train)

y_pred_et = et_model.predict(X_test)
```

# Gradient Boost Implementation

```python
from sklearn.ensemble import GradientBoostingClassifier

gb_model = GradientBoostingClassifier(random_state=11)
gb_model.fit(X_train, y_train)

y_pred_gb= gb_model.predict(X_test)
```

# XGBoost Implementation

```python
#X_train, X_test, y_numeric_train, y_numeric_test = train_test_split(
  #  X, y_numeric, test_size=0.2, random_state=11, stratify=y_numeric
#)
from xgboost import XGBClassifier

xgb_model =XGBClassifier(
    n_estimators=100,
    learning_rate=0.1,
    random_state=11,
    eval_metric='mlogloss'  # Add for multiclass
)

xgb_model.fit(X_train, y_train_numeric)
y_pred_xgb = xgb_model.predict(X_test)
y_pred_xgb_labels = le.inverse_transform(y_pred_xgb)  # Convert back for evaluation
```

# SVM implementation

```python
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
model_svm1=SVC(kernel='linear',C=1.0,random_state=11,probability=True)
model_svm2=SVC(kernel='rbf',C=1.0,random_state=11,probability=True)
model_svm3=SVC(kernel='poly',C=1.0,random_state=11,probability=True)
model_svm1.fit(X_train_scaled,y_train)
model_svm2.fit(X_train_scaled,y_train)
model_svm3.fit(X_train_scaled,y_train)
y_pred_svm1=model_svm1.predict(X_test_scaled)
y_pred_svm2=model_svm2.predict(X_test_scaled)
y_pred_svm3=model_svm3.predict(X_test_scaled)
```

# Logistic Regression Implementation

```python
from sklearn.linear_model import LogisticRegression
model_lr=LogisticRegression()
model_lr.fit(X_train,y_train)
y_pred_lr=model_lr.predict(X_test)
```

```python
# 4. Check test set size
print(f"Training samples: {X_train.shape[0]}")
print(f"Testing samples: {X_test.shape[0]}")
print(f"Unique diseases in test: {y_test.nunique()}")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import pandas as pd

results = []

def evaluate_model(name, y_test, y_pred):
    results.append({
        "Model": name,
        "Accuracy": accuracy_score(y_test, y_pred),
        "Precision": precision_score(y_test, y_pred, average='weighted'),
        "Recall": recall_score(y_test, y_pred, average='weighted'),
        "F1 Score": f1_score(y_test, y_pred, average='weighted')
    })

# Add all models
evaluate_model("Decision Tree", y_test, y_pred_dt)
evaluate_model("Random Forest", y_test, y_pred_rf)
evaluate_model("Extremely Randomized Trees", y_test, y_pred_et)
evaluate_model("Gradient Boosting", y_test, y_pred_gb)
evaluate_model("XGradient Boosting", y_test, y_pred_xgb_labels)
evaluate_model("SVM1", y_test, y_pred_svm1)
evaluate_model("SVM2", y_test, y_pred_svm2)
evaluate_model("SVM3", y_test, y_pred_svm3)
evaluate_model("Logistic Regression", y_test, y_pred_lr)
# Create DataFrame
comparison_df = pd.DataFrame(results)

# Sort by Accuracy
comparison_df = comparison_df.sort_values(by="Accuracy", ascending=False).reset_index(drop=True)

print(comparison_df)

# =========================
# Best Model Selection ⭐
# =========================

best_model = comparison_df.iloc[0]

print("\nBest Model:")
print(best_model)
```

```python
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.metrics import make_scorer, accuracy_score
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Prepare data (use full dataset for cross-validation)
X = df.drop('prognosis', axis=1)
y = df['prognosis']

# Define cross-validation strategy (5-fold stratified)
cv_strategy = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Store results
cv_results = []

# Models to evaluate
models = {
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
    "Extremely Randomized Trees": ExtraTreesClassifier(n_estimators=100, random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(random_state=42),
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "XGBoost": XGBClassifier(n_estimators=100, learning_rate=0.1, random_state=42, eval_metric='mlogloss')
}

# For SVM (needs scaling, so we'll do pipeline)
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

svm_models = {
    "SVM Linear": Pipeline([('scaler', StandardScaler()), ('svm', SVC(kernel='linear', random_state=42))]),
    "SVM RBF": Pipeline([('scaler', StandardScaler()), ('svm', SVC(kernel='rbf', random_state=42))]),
    "SVM Poly": Pipeline([('scaler', StandardScaler()), ('svm', SVC(kernel='poly', random_state=42))])
}

# Add SVM models to dictionary
models.update(svm_models)

print("="*80)
print("CROSS-VALIDATION RESULTS (5-Fold Stratified)")
print("="*80)
print()

# Perform cross-validation for each model
for name, model in models.items():
    # For XGBoost, we need encoded labels
    if name == "XGBoost":
        le = LabelEncoder()
        y_encoded = le.fit_transform(y)
        cv_scores = cross_val_score(model, X, y_encoded, cv=cv_strategy, scoring='accuracy')
        cv_scores_mean = cv_scores.mean()
        cv_scores_std = cv_scores.std()
    else:
        cv_scores = cross_val_score(model, X, y, cv=cv_strategy, scoring='accuracy')
        cv_scores_mean = cv_scores.mean()
        cv_scores_std = cv_scores.std()
    
    # Calculate confidence interval
    ci_lower = cv_scores_mean - 1.96 * cv_scores_std
    ci_upper = cv_scores_mean + 1.96 * cv_scores_std
    
    cv_results.append({
        "Model": name,
        "CV Mean Accuracy": cv_scores_mean,
        "CV Std Dev": cv_scores_std,
        "95% CI Lower": ci_lower,
        "95% CI Upper": ci_upper,
        "Individual Scores": cv_scores
    })
    
    print(f"📊 {name}:")
    print(f"   Mean Accuracy: {cv_scores_mean:.4f} (+/- {cv_scores_std:.4f})")
    print(f"   95% Confidence Interval: [{ci_lower:.4f}, {ci_upper:.4f}]")
    print(f"   Fold Scores: {cv_scores}")
    print()

# Create results DataFrame
cv_results_df = pd.DataFrame(cv_results)
cv_results_df = cv_results_df.sort_values(by="CV Mean Accuracy", ascending=False).reset_index(drop=True)

print("="*80)
print("RANKING BY CROSS-VALIDATION ACCURACY")
print("="*80)
print(cv_results_df[["Model", "CV Mean Accuracy", "CV Std Dev"]].to_string(index=False))

# Find best model
best_model_cv = cv_results_df.iloc[0]

print("\n" + "="*80)
print("🏆 BEST MODEL BASED ON CROSS-VALIDATION")
print("="*80)
print(f"\n✅ Best Model: {best_model_cv['Model']}")
print(f"   ├─ Cross-Validation Accuracy: {best_model_cv['CV Mean Accuracy']:.4f} (+/- {best_model_cv['CV Std Dev']:.4f})")
print(f"   └─ 95% Confidence Interval: [{best_model_cv['95% CI Lower']:.4f}, {best_model_cv['95% CI Upper']:.4f}]")

# Statistical significance test (compare best vs second best)
if len(cv_results_df) > 1:
    second_best = cv_results_df.iloc[1]
    print(f"\n📊 Comparison with second best ({second_best['Model']}):")
    print(f"   Difference: {best_model_cv['CV Mean Accuracy'] - second_best['CV Mean Accuracy']:.4f}")
    
    # Simple t-test approximation
    if (best_model_cv['CV Mean Accuracy'] - second_best['CV Mean Accuracy']) > 2 * max(best_model_cv['CV Std Dev'], second_best['CV Std Dev']):
        print("   ✅ Statistically significant difference")
    else:
        print("   ⚠️ Difference may not be statistically significant")

# Train final best model on all data
print("\n" + "="*80)
print("TRAINING FINAL BEST MODEL")
print("="*80)

if best_model_cv['Model'] == "XGBoost":
    le = LabelEncoder()
    y_encoded = le.fit_transform(y)
    best_final_model = models[best_model_cv['Model']].fit(X, y_encoded)
    print(f"✅ Trained {best_model_cv['Model']} on all {len(X)} samples")
    print(f"   Model can now predict diseases for new symptoms")
else:
    best_final_model = models[best_model_cv['Model']].fit(X, y)
    print(f"✅ Trained {best_model_cv['Model']} on all {len(X)} samples")
    print(f"   Model can now predict diseases for new symptoms")

# Visual comparison
print("\n" + "="*80)
print("VISUAL COMPARISON")
print("="*80)

plt.figure(figsize=(12, 6))
models_names = cv_results_df['Model'].values
means = cv_results_df['CV Mean Accuracy'].values
stds = cv_results_df['CV Std Dev'].values

# Create horizontal bar chart
bars = plt.barh(range(len(models_names)), means, xerr=stds, capsize=5, alpha=0.7, color='steelblue')
plt.yticks(range(len(models_names)), models_names)
plt.xlabel('Cross-Validation Accuracy')
plt.title('Model Performance Comparison with 95% Confidence Intervals')
plt.xlim(0, 1.05)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Find models with no overfitting (small standard deviation)
stable_models = cv_results_df[cv_results_df['CV Std Dev'] < 0.02]

print("\n" + "="*80)
print("🎯 FINAL RECOMMENDATION")
print("="*80)
print(f"""
Based on 5-fold stratified cross-validation:

🏆 USE: {best_model_cv['Model']}
   ├─ Expected Accuracy: {best_model_cv['CV Mean Accuracy']:.2%}
   ├─ Stability: {'High' if best_model_cv['CV Std Dev'] < 0.02 else 'Medium'}
   ├─ 95% Confidence: [{best_model_cv['95% CI Lower']:.2%}, {best_model_cv['95% CI Upper']:.2%}]
   └─ Reliable for: Real-world disease prediction

💡 Why this is better than your previous evaluation:
   • Cross-validation uses all data (not just 42 test samples)
   • Provides realistic performance estimates
   • Detects if models are overfitting to specific data splits
""")

# Print top 3 models
print("\n📊 TOP 3 MODELS:")
for i in range(min(3, len(cv_results_df))):
    model = cv_results_df.iloc[i]
    print(f"   {i+1}. {model['Model']}: {model['CV Mean Accuracy']:.4f} (±{model['CV Std Dev']:.4f})")
```
